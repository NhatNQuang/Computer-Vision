```python
img = Image.open("C:\\Users\\nguye\\OneDrive\\Documents\\Nguyễn Quang Nhật\\Quang Nhat\\IMG_1043.JPG")
```

```python
gray = img.convert('L')

# Change image into Numpy array
arr = np.array(gray)
```

```python
# Biến đổi ảnh xám theo yêu cầu
# Chuyển độ xám yếu thành mạnh, và ngược lại
invert_img = 255 - arr
plt.imshow(invert_img, cmap='gray')
```

```python
# Bình phương giá trị pixel
squared_img = arr ** 2
plt.imshow(squared_img, cmap='gray')
# Giowis hạn giá trị pixel trong khoảng 100-200
clampled_img = np.clip(arr, 100, 200)
plt.imshow(clampled_img, cmap='gray')
```

```python
# Hiển thị hình ảnh gốc và hình ảnh biến đổi với khoảng trắng
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(arr, cmap='gray')
axs[0, 0].set_title('Hình ảnh gốc')

axs[0, 1].imshow(invert_img, cmap='gray')
axs[0, 1].set_title('Hình ảnh đã biến đổi')

axs[1, 0].imshow(squared_img, cmap='gray')
axs[1, 0].set_title('Hình ảnh đã biến đổi')

axs[1, 1].imshow(clampled_img, cmap='gray')
axs[1, 1].set_title('Hình ảnh đã biến đổi')

plt.show()
```
