LINE CHART

```python
x = [100, 200, 300, 400, 500]
y = [100, 160, 200, 300, 250]

x1 = [170, 190, 310, 380, 440]
y1 = [120, 180, 280, 330, 390]
plt.plot(x, y, 'bo-', label="data")
plt.plot(x1, y1, marker='*', color = 'red', linestyle=':', label="number")
plt.legend() #show label

# Set title
plt.title('Line chart')
# Set label for x, y
plt.xlabel(' x axis')
plt.ylabel('y axis')
plt.show()
```

BAR CHART

```python
categories = ['A', 'B', 'C']
values = [15, 10, 25]

plt.bar(categories,values, color='red', alpha=0.8)# the strength of color
# Set title
plt.title('Bar chart')
# Set label for x, y
plt.xlabel(' x axis')
plt.ylabel('y axis')
plt.show()
```

SHOW THE WORD CLOUD of A TEXT

```python
from wordcloud import WordCloud

text = "Hôm nay và trời đẹp quá, đẹp và đẹp quá, trời mưa và rất to, to cực kì "

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear') # interpolation: làm mịn hình ảnh cho đẹp khi hiển thị
plt.axis('off')
plt.title('biểu đồ đám mây từ văn bản ví dụ')
plt.show()
```

RADAR CHART

```python
categories = ['A', 'B', 'C','D','E']
values = [2, 6, 4, 7, 5]

N = len(categories)
values+= values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Bieu do radar')
plt.show()
```
