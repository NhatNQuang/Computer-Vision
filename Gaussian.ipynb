Bộ lọc Gaussian làm mờ ảnh

Làm mờ hình ảnh bằng cách giảm độ biến đổi củ màu sắc giữa các điểm ảnh gần nhau

Hữu ích với 2D

```python
from library import *
```

```python
img = Image.open("C:\\Users\\nguye\\OneDrive\\Documents\\Nguyễn Quang Nhật\\Quang Nhat\\chan dung.jpg")

# Anhr xám hoặc không
gray_image = img.convert('L')

# Chuyển ảnh xám về mảng
img_a = np.array(gray_image)

# Phải sử dụng ndimage
# Làm mờ ảnh xám với bộ lọc Gaussian
sigma1 = 0.5
blurred_image1 = ndimage.gaussian_filter(img_a, sigma=0.5, mode="constant")
blurred_image2 = ndimage.gaussian_filter(img_a, sigma=3, mode="constant")
blurred_image3 = ndimage.gaussian_filter(img_a, sigma=7, mode="constant")

plt.figure(figsize=(15, 10))

# Hiển thị ảnh
plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Anhr xám gốc')

plt.subplot(2,2,2)
plt.imshow(blurred_image1, cmap='gray')
plt.title('sigma=0.5')

plt.subplot(2,2,3)
plt.imshow(blurred_image2, cmap='gray')
plt.title('sigma=3')

plt.subplot(2,2,4)
plt.imshow(blurred_image3, cmap='gray')
plt.title('sigma=7')
```
