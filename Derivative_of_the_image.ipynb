# BỘ LỌC SOBEL

Bộ lọc Sobel sử dụng 2 mặt nạ( Mặt nạ Sobel x và Sobel y) để tính đọa hàm riêng theo hướng NGANG và DỌC.

=>> Dùng để phát hiện cạnh trong ảnh

```python
img = Image.open("C:\\Users\\nguye\\OneDrive\\Documents\\Nguyễn Quang Nhật\\Quang Nhat\\chan dung.jpg")

# Anhr xám hoặc không
gray_image = img.convert('L')

# Chuyển ảnh xám về mảng
img_a = np.array(gray_image)

# Tính gradient theo hướng x
gradient_x = ndimage.sobel(img_a, axis=0, mode='constant')
gradient_y = ndimage.sobel(img_a, axis=1, mode='constant')


# Hiển thị ảnh
fig, axs = plt.subplots(2,2, figsize=(10, 8))


plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('ảnh xám gốc')

plt.subplot(2,2,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('ảnh gradient theo hướng x')

plt.subplot(2,2,3)
plt.imshow(gradient_y, cmap='gray')
plt.title('ảnh gradient theo hướng y')

plt.show()
```

# Bộ lọc Prewitt

Phát hiện biên và cạnh trong ảnh.

Bộ lọc này được thiết kế để tìm sự biến đổi của cường độ màu sắc tại mỗi điểm ảnh trong hình ảnh


```python
img = Image.open("C:\\Users\\nguye\\OneDrive\\Documents\\Nguyễn Quang Nhật\\Quang Nhat\\chan dung.jpg")

# Anhr xám hoặc không
gray_image = img.convert('L')

# Chuyển ảnh xám về mảng
img_a = np.array(gray_image)

# Tính gradient theo hướng x (đạo hàm riêng theo x)
gradient_x = ndimage.convolve(img_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tính gradient theo hướng x (đạo hàm riêng theo y)
gradient_y = ndimage.convolve(img_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tính biên độ gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hiển thị ảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))


plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('ảnh xám gốc')

plt.subplot(2,1,2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('ảnh gradient')


plt.show()
```
